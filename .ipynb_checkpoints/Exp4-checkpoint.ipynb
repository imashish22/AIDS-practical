```python
import pandas as pd
realEstate_dataset = pd.read_csv('realestate.csv')
print(realEstate_dataset.head())
```

```python
realEstate_dataset_copy = realEstate_dataset
mean_age = realEstate_dataset_copy['house_size'].mean()
std_age = realEstate_dataset_copy['house_size'].std()
realEstate_dataset_copy['house_size_normalized'] = (realEstate_dataset['house_size'] - mean_age) / std_age
print(realEstate_dataset_copy[['house_size', 'house_size_normalized']].head())
```

```python
import matplotlib.pyplot as plt
# Plot histograms of the original 'house_size' attribute and the normalized 'house_size_normalized' attribute
plt.figure(figsize=(12, 6))
# Plot histogram of the original 'house_size' attribute
plt.subplot(1, 2, 1)
plt.hist(realEstate_dataset_copy['house_size'], bins=20, color='blue', alpha=0.7)
plt.title('Histogram of house_size')
plt.xlabel('house_size')
plt.ylabel('Frequency')
# Plot histogram of the normalized 'house_size_normalized' attribute
plt.subplot(1, 2, 2)
plt.hist(realEstate_dataset_copy['house_size_normalized'], bins=20, color='green', alpha=0.7)
plt.title('Histogram of house_size (Normalized)')
plt.xlabel('Normalized house_size')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
